Sentiment Funtion 1 = TextBlob

```python
from textblob import TextBlob
import re
import string
from nltk.corpus import stopwords

# Download stopwords if not available
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

def clean_text(text):
    """Remove punctuation, numbers, and stopwords from the text."""
    text = text.lower()
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.translate(str.maketrans("", "", string.punctuation))  # Remove punctuation
    words = text.split()
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    return " ".join(words)

def get_sentiment(text):
    """Return sentiment polarity as 'positive', 'negative', or 'neutral'."""
    cleaned_text = clean_text(text)
    sentiment_score = TextBlob(cleaned_text).sentiment.polarity  # Polarity (-1 to 1)

    if sentiment_score > 0:
        return "positive"
    elif sentiment_score < 0:
        return "negative"
    else:
        return "neutral"


# Define expected sentiment based on rating
df["expected_sentiment"] = df["rating"].apply(lambda x: "positive" if x >= 4 else "negative" if x <= 2 else "neutral")

```

Sentiment Funtion 2: Vader

```python

from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download("vader_lexicon")
sia = SentimentIntensityAnalyzer()

def get_vader_sentiment(text):
    """
    Returns sentiment as 'positive', 'negative', or 'neutral' using VADER.
    """
    sentiment_score = sia.polarity_scores(text)["compound"]  # Compound sentiment score (-1 to 1)

    if sentiment_score > 0.05:
        return "positive"
    elif sentiment_score < -0.05:
        return "negative"
    else:
        return "neutral"


```

CRITERION 2 : FOR SENTIMENT, APPLY BOTH FUNTIONS, TextBlob + Vader

```python
# Apply sentiment analysis using both functions
df["sentiment_textblob"] = df["title"].apply(get_sentiment)  # First method (TextBlob or similar)
df["sentiment_vader"] = df["title"].apply(get_vader_sentiment)  # VADER method

# Determine final sentiment: Negative if either method predicts negative
df["final_review_sentiment"] = df.apply(
    lambda row: "negative" if "negative" in [row["sentiment_textblob"], row["sentiment_vader"]]
    else "positive" if "positive" in [row["sentiment_textblob"], row["sentiment_vader"]]
    else "neutral",
    axis=1
)

#"Label" by comparing "Expected" sentiment with "Final" sentiment
# Filter reviews where:
# - Expected: Positive, but detected as Negative
# - Expected: Negative, but detected as Positive
# - Exclude cases where both predicted and expected are neutral
sentiment_conflicts = df[
    ((df["expected_sentiment"] == "positive") & (df["final_review_sentiment"] == "negative")) |
    ((df["expected_sentiment"] == "negative") & (df["final_review_sentiment"] == "positive"))
]

# Print first 20 mismatched cases
print(sentiment_conflicts[["rating", "title", "sentiment_textblob", "sentiment_vader", "final_review_sentiment", "expected_sentiment"]].head(20))
```

Labeling as fake on conflicting sentiment

```python
df.loc[
    ((df["expected_sentiment"] == "positive") & (df["final_review_sentiment"] == "negative")) |
    ((df["expected_sentiment"] == "negative") & (df["final_review_sentiment"] == "positive")),
    "fake"
] = 1
```

CRITERION 3: Labeling as fake to user who reviewed 20+ time within these 400000 reviews

```python
# Count number of reviews per user
user_review_counts = df['user_id'].value_counts()

# Define a threshold (e.g., more than 20 reviews)
suspicious_users = user_review_counts[user_review_counts > 20].index

# Mark reviews from suspicious users as fake (1)
df.loc[df['user_id'].isin(suspicious_users), 'fake'] = 1

```

After applying above 3 criterions, now the 4th criterion
CRITERION 4: If the expected sentiment matches the final review sentiment or the final sentiment is neutral, and the rating is between 2 and 4, then mark the review as non-fake 0.

```python
df.loc[
    ((df["expected_sentiment"] == df["final_review_sentiment"]) | (df["final_review_sentiment"] == "neutral") | (df['expected_sentiment'] == "neutral")) & 
    (df["rating"].isin([2, 4])), 
    "fake"
] = 0

df.loc[df["rating"] == 3, "fake"] = 0
```
